```python
import pandas as pd
import matplotlib.pyplot as plt

# Importing the data
data = pd.read_excel("C:/Users/Olena/CompareNew.xlsx")
print(data.columns)

num_nodes = data['Number of nodes']

# Getting the time of working of each algorithm
time_kk = data['KK     time, s']
time_kknn = data['KKNN time, s']
time_neulay = data['NeuLay-2 time, s']

# Creating the plot
plt.figure(figsize=(10, 6))
plt.rcParams.update({'font.size': 18})
plt.plot(num_nodes, time_kk, 'ro', label='KK')
plt.plot(num_nodes, time_kknn, 'go', label='KKNN')
plt.plot(num_nodes, time_neulay, 'bo', label='NeuLay-2')

plt.xlabel('Number of nodes')
plt.ylabel('Time, s')
plt.legend()

# Drawing the plot
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Importing the data
data = pd.read_excel("C:/Users/Olena/CompareNew.xlsx")

num_nodes = data['Number of nodes']

# Getting the KKNN/KK values
time_algorithms = data['KKNN/KK']

graph_names = data['Graph name']

colors = ['purple']*2 + ['blue']*5 + ['red']*3 + ['green']*4 + ['black']*6

plt.figure(figsize=(12, 8))
plt.rcParams.update({'font.size': 18})

# Adding the names
for i, color in enumerate(colors):
    plt.plot(num_nodes[i], time_algorithms[i], 'o', color=color)
    if (i==1) or (i==4) or (i==9) or (i==11):
        plt.text(num_nodes[i], time_algorithms[i] - 0.1, graph_names[i], ha='center', va='bottom', fontsize=11, color=color)
    elif (i==17):
        plt.text(num_nodes[i]-125, time_algorithms[i]-0.05, graph_names[i], ha='center', va='bottom', fontsize=11, color=color)
    elif (i==13) or (i==14):
        plt.text(num_nodes[i]-125, time_algorithms[i]-0.09, graph_names[i], ha='center', va='bottom', fontsize=11, color=color)
    elif (i==2) or (i==3):
        plt.text(num_nodes[i]+200, time_algorithms[i] + 0.02, graph_names[i], ha='center', va='bottom', fontsize=11, color=color)
    elif (i==7):
        plt.text(num_nodes[i]+150, time_algorithms[i]-0.05, graph_names[i], ha='center', va='bottom', fontsize=11, color=color)
    else:
        plt.text(num_nodes[i], time_algorithms[i] + 0.04, graph_names[i], ha='center', va='bottom', fontsize=11, color=color)

plt.axhline(y=1, color='red', linestyle='--', label='Threshold = 1')
plt.xlabel('Number of nodes')
plt.ylabel('timeKKNN/timeKK')
plt.scatter([], [], color='purple', label='SmallWorld')
plt.scatter([], [], color='blue', label='Internet')
plt.scatter([], [], color='red', label='ScaleFree')
plt.scatter([], [], color='green', label='Geometric')
plt.scatter([], [], color='black', label='Symmetric')
plt.legend(loc='upper right', fontsize='large')
plt.grid(True)
```
